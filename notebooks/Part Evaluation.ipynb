```python
import sys
sys.path.append('../src/')
sys.path.append('/home/ipl/installs/caffe-rc/python/')
from matplotlib import pylab as plt
from datetime import datetime as dt
import numpy as np
import caffe
import cv2

from deep_extractor import CNN_Features_CAFFE_REFERENCE
from dataset import CUB_200_2011
from storage import datastore
import utils
import settings
import parts
import cub_utils
import rects

%matplotlib inline
```

```python
cub = CUB_200_2011(settings.CUB_ROOT)
all_image_infos = cub.get_all_image_infos()
IDtrain, IDtest = cub.get_train_test_id()
cub_parts = cub.get_parts()
dh = cub_utils.DeepHelper()
bah = cub_utils.BerkeleyAnnotationsHelper(settings.BERKELEY_ANNOTATION_BASE_PATH, IDtrain, IDtest)
bboxes = cub.get_bbox()

BRGh = rects.BerkeleyRG(settings.BERKELEY_ANNOTATION_BASE_PATH, IDtrain, IDtest, 'head')
BRGh.setup()

BRGb = rects.BerkeleyRG(settings.BERKELEY_ANNOTATION_BASE_PATH, IDtrain, IDtest, 'bbox')
BRGb.setup()
```

```python
img_id = 7415

fig = plt.figure(figsize=(15, 10))
ax1 = fig.add_subplot(131)
ax2 = fig.add_subplot(132)
ax3 = fig.add_subplot(133)

img = caffe.io.load_image(all_image_infos[img_id])
img_r = cv2.resize(img, (227, 227))
img_parts = cub_parts.for_image(img_id)
img_bbox = bboxes[img_id - 1]

rect_head = BRGh.generate(img_id)
rect_bbox = BRGb.generate(img_id)

only_body = rect_bbox.get_rect(img)

rect_head.norm_for_bbox(img_bbox)
only_body = rect_head.draw_rect(only_body, color=(1, 1, 0))
rect_head.denorm_for_bbox(img_bbox)


rect_head.norm_for_size(img.shape, 227)
rect_bbox.norm_for_size(img.shape, 227)
img_r = rect_head.draw_rect(img_r, color=(1, 1, 0))
img_r = rect_bbox.draw_rect(img_r)

rect_head.denorm_for_size(img.shape, 227)
rect_bbox.denorm_for_size(img.shape, 227)

img = rect_head.draw_rect(img, color=(1, 1, 0))
img = rect_bbox.draw_rect(img)

ax1.imshow(img)
ax2.imshow(only_body)
ax3.imshow(img_r)
```
