### computing ys

```python
def convert_to_y(orc_bbs, est_bbs):
    y = []
    for i in range(len(orc_bbs)):
        if not orc_bbs[i].is_invalid():
            Gx, Gy, Gw, Gh = orc_bbs[i]._get_cendim()
            Px, Py, Pw, Ph = est_bbs[i]._get_cendim()
            tx = (Gx - Px) / float(Pw)
            ty = (Gy - Py) / float(Ph)
            tw = np.log(Gw / float(Pw))
            th = np.log(Gh / float(Ph))
            y.append([tx, ty, tw, th])
        else:
            print i
    return np.array(y)
```
