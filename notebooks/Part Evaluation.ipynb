```python
cub = CUB_200_2011(settings.CUB_ROOT)
all_image_infos = cub.get_all_image_infos()
IDtrain, IDtest = cub.get_train_test_id()
cub_parts = cub.get_parts()
dh = cub_utils.DeepHelper()
bah = cub_utils.BerkeleyAnnotationsHelper(settings.BERKELEY_ANNOTATION_BASE_PATH, IDtrain, IDtest)
bboxes = cub.get_bbox()

BRGh = rects.BerkeleyRG(settings.BERKELEY_ANNOTATION_BASE_PATH, IDtrain, IDtest, 'head')
BRGh.setup()

BRGb = rects.BerkeleyRG(settings.BERKELEY_ANNOTATION_BASE_PATH, IDtrain, IDtest, 'bbox')
BRGb.setup()

SRGh = rects.SharifRG(cub, 'head', alpha=0.5)
SRGh.setup()
```

```python
img_id = 9

fig = plt.figure(figsize=(15, 10))
ax1 = fig.add_subplot(111)

img = caffe.io.load_image(all_image_infos[img_id])
img_r = cv2.resize(img, (227, 227))
img_parts = cub_parts.for_image(img_id)
img_bbox = bboxes[img_id - 1]

rect_head_b = BRGh.generate(img_id)
rect_head_s = SRGh.generate(img_id, img.shape)
rect_head_s.add_noise(center_std=5, dimension_std=5)

img = rect_head_b.draw_rect(img, color=(1, 1, 0))
img = rect_head_s.draw_rect(img, color=(0, 1, 1))

print 'before cropping'
print rect_head_s.evalIOU(rect_head_b, img.shape)

ax1.imshow(img)
```

```python
S = 0
for t in cub.get_all_images():
    img_id = int(t['img_id'])
    img_addr = t['img_file']
    
    img = cv2.imread(img_addr)
    
    bgt = BRGh.generate(img_id)
    sgt = SRGh.generate(img_id, img.shape)
    
    pcp_score = sgt.evalPCP(bgt, img.shape, thresh=0.75)
    S += pcp_score
    
#     if pcp_score == 0:
#         print img_id

print S
```
