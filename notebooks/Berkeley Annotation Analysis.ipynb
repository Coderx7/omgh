```python
cub = CUB_200_2011(settings.CUB_ROOT)
all_image_infos = cub.get_all_image_infos()
IDtrain, IDtest = cub.get_train_test_id()
bbox = cub.get_bbox()
cub_parts = cub.get_parts()
```

```python
b_train_anno = scipy.io.loadmat('/home/ipl/repo/part-based-RCNN/annotations/bird_train.mat')
b_train_anno = b_train_anno['data']

b_test_anno = scipy.io.loadmat('/home/ipl/repo/part-based-RCNN/annotations/bird_test.mat')
b_test_anno = b_test_anno['data']
```

```python
def get_train_berkeley_annotation(train_id, name):
    p = 0
    if name == 'head':
        p = 1
    elif name == 'body':
        p = 2
    elif name == 'bbox':
        p = 3
    res = b_train_anno[0, train_id][p][0]
    ymin, xmin, ymax, xmax= res[0], res[1], res[2], res[3]
    
    return xmin, xmax, ymin, ymax

def get_test_berkeley_annotation(test_id, name):
    p = 0
    if name == 'bbox':
        p = 1
    elif name == 'head':
        p = 2
    elif name == 'body':
        p = 3
    res = b_test_anno[0, test_id][p][0]
    ymin, xmin, ymax, xmax= res[0], res[1], res[2], res[3]
    
    return xmin, xmax, ymin, ymax

def get_berkeley_annotation(img_id, name):
    train_where = np.argwhere(IDtrain == img_id)
    test_where = np.argwhere(IDtest == img_id)
    if train_where.shape[0] == 1:
        return get_train_berkeley_annotation(train_where[0, 0], name)
    elif test_where.shape[0] == 1:
        return get_test_berkeley_annotation(test_where[0, 0], name)
    else:
        raise Exception('Not found of multiple found!')
        
```

# testing oracle performance without fine-tuning

```python
fe_r = CNN_Features_CAFFE_REFERENCE(datastore(settings.storage('ccr')))
fe_c = CNN_Features_CAFFE_REFERENCE(datastore(settings.storage('ccc')))
fe_h = CNN_Features_CAFFE_REFERENCE(datastore(settings.storage('ccphead')))
fe_b = CNN_Features_CAFFE_REFERENCE(datastore(settings.storage('ccpbody')))

Xtrain_r, ytrain_r, Xtest_r, ytest_r = cub.get_train_test(fe_r.extract_one)
Xtrain_c, ytrain_c, Xtest_c, ytest_c = cub.get_train_test(fe_c.extract_one)
Xtrain_h, ytrain_h, Xtest_h, ytest_h = cub.get_train_test(fe_h.extract_one)
Xtrain_b, ytrain_b, Xtest_b, ytest_b = cub.get_train_test(fe_b.extract_one)
```

```python
ytrain = ytrain_r
ytest = ytest_r

Xtrain = np.concatenate((Xtrain_r, Xtrain_c, Xtrain_h, Xtrain_b), axis=1)
Xtest = np.concatenate((Xtest_r, Xtest_c, Xtest_h, Xtest_b), axis=1)
```

```python
model = sklearn.svm.LinearSVC(C=0.0001)
model.fit(Xtrain, ytrain)
preds = model.predict(Xtest)

print utils.mean_accuracy(ytest, preds)
```
