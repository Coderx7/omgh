```python
import caffe
import numpy as np
from matplotlib import pylab as plt
%matplotlib inline
```

```python
model_file = '/home/ipl/installs/caffe/models/cub_vgg_whole/deploy.prototxt'
pretrained = '/home/ipl/installs/caffe/models/cub_vgg_whole/cub_iter_80000.caffemodel'

caffe.set_mode_gpu()
mean_pix = np.array([103.939, 116.779, 123.68])

net = caffe.Classifier(model_file, pretrained,
                       mean=mean_pix,
                       channel_swap=(2,1,0),
                       raw_scale=255)
```

```python
img = caffe.io.load_image(
    '/home/ipl/datasets/CUB-200-2011/CUB_200_2011/CUB_200_2011/images/051.Horned_Grebe/Horned_Grebe_0078_34799.jpg')
plt.imshow(img)
print img.shape
```

```python
preds = net.predict([img], oversample=False)

plt.plot(preds.flatten())

print np.argmax(preds.flatten())
```
