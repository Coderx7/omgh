```python
from parts import *
cub = CUB_200_2011(settings.CUB_ROOT)
cub_parts = cub.get_parts()

def vis_near(test_id, train_id):
    img_test = caffe.io.load_image(cub.get_image_info(test_id))
    parts_test = cub_parts.for_image(test_id)
    
    img_train = caffe.io.load_image(cub.get_image_info(train_id))
    parts_train = cub_parts.for_image(train_id)
    
    fig = plt.figure(figsize=(15, 10))
    
    ax = fig.add_subplot(121)
    # ax.set_axis_off()
    filtered_parts_test = parts_test.filter_by_name(Parts.HEAD_PART_NAMES)
    filtered_parts_test.draw_rect(img_test)
    ax.imshow(img_test)
    ax.set_title(test_id)
    ax.set_xlim((0, img_test.shape[1]))
    ax.set_ylim((img_test.shape[0], 0))
    parts_test.draw_part(ax)
    
    ax = fig.add_subplot(122)
    # ax.set_axis_off()
    filtered_parts_train = parts_train.filter_by_name(Parts.HEAD_PART_NAMES)
    filtered_parts_train.draw_rect(img_train)
    ax.imshow(img_train)
    ax.set_title(train_id)
    ax.set_xlim((0, img_train.shape[1]))
    ax.set_ylim((img_train.shape[0], 0))
    parts_train.draw_part(ax)
```

```python
tests_id = [201, 205]
train_ids = [805, 805]

for i, j in zip(tests_id, train_ids):
    vis_near(i, j)
```
